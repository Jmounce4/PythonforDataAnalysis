```python
import pandas as pd
import seaborn as sns
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt
```

```python
def binomial(attempts, successRate, successes):

    #graph

    p = binom.pmf(list(range(0, attempts+1, 1)), n=attempts, p=successRate)
    df1 = pd.DataFrame(
        {'probability': p,
         'amount': np.arange(0, attempts+1, 1),
        })
    fig = sns.barplot(x='amount', y='probability', data=df1, hue=df1.probability, palette='coolwarm')
    plt.figure(figsize=(15,15))
    sns.despine()
    plt.show()


    #statement
    p2 = binom.pmf(successes, n=attempts, p=successRate)
    percentage = p2*100
    fractional = 1/p2
    rounded_fractional = round(fractional)
    
    print('The chance of getting ', successes, ' is', f"{percentage:.5f}%")
    print('or 1 in', rounded_fractional)
    return
```

```python
#attempts, rate, amount wanted
binomial(20, 0.2, 10)
```
